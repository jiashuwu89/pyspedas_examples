from pyspedas.geopack import get_tsy_params
from pyspedas import time_double

Generate the expected field using the T96 model at the THEMIS-A times and GSM positions, using .

 Plot the modeled field
 (Hint: position variable + "_bt96") 

````python
import pyspedas
from pyspedas.geopack import tt96, get_tsy_params
from pyspedas import tplot,join_vec, time_double

def get_params(trange, position_var):
    support_trange = [time_double(trange[0])-60*60*24, 
                      time_double(trange[1])+60*60*24]
    pyspedas.kyoto.dst(trange=support_trange)
    pyspedas.omni.data(trange=support_trange)
    join_vec(['BX_GSE', 'BY_GSM', 'BZ_GSM'])
    parmod = get_tsy_params('kyoto_dst',
                    'BX_GSE-BY_GSM-BZ_GSM_joined',
                    'proton_density',
                    'flow_speed',
                    model='t96',
                    pressure_tvar='Pressure',
                    g_variables=None,
                    speed=True)
    # The field models expect the model parameters to be supplied at the same timestamps
    # as the input positions, so we interpolate them with pyspedas.tinterpol()
    pyspedas.tinterpol(parmod,position_var,newname='parmod_interpolated')
    return 'parmod_interpolated'

state_vars = pyspedas.themis.state(probe='a',trange=["2023-03-24","2023-03-25"],varnames=["tha_pos_gsm"])
parmod = get_params(["2023-03-24","2023-03-25"], 'tha_pos_gsm')
print(state_vars)
tt96('tha_pos_gsm',parmod)
tplot('tha_pos_gsm_bt96')
````

Load THEMIS state data for probe A, 2023-03-24, in GSM coordinates

Generate the expected field using the T89 model at the THEMIS-A times and GSM positions, using the default model parameter.

Hint: 

```

```
 
Output: position_variable + '_bt89'

 
Load THEMIS-A FGM spin fit data for the same time range 

Hint: 
```pyspedas.themis.fgm```

output variable 'tha_fgs_gsm'

Calculate the difference between the observed and modeled field variables.
(Hint: ```pyspedas.subtract```)

Set the yrange of the difference variable to [-200.0, 200.0] nT

Hint: `pyspedas.options`, `"yrange"`

Plot the position variable, modeled field variable, observed field variable, and observed-modeled field difference.
 


````python
import pyspedas
from pyspedas.geopack import tt89
from pyspedas import tplot, subtract, options

trange = ["2023-03-24","2023-03-25"]
state_vars = pyspedas.themis.state(probe='a',trange=trange,varnames=["tha_pos_gsm"])
print(state_vars)
tt89('tha_pos_gsm')

fgm_vars = pyspedas.themis.fgm(probe='a',trange=trange)

subtract('tha_fgs_gsm','tha_pos_gsm_bt89',newname='mag_diff')

options('mag_diff', 'yrange', [-200.0, 200.0])
options('mag_diff', 'ytitle', 'Obs-T89')

tplot(["tha_pos_gsm","tha_pos_gsm_bt89", "tha_fgs_gsm", "mag_diff"])
````

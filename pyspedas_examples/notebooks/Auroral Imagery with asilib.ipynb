This notebook shows examples of how to use M. Shumko's asilib library to download keogram and all-sky imager data, and display as fisheye images, images projected on maps, and movies.

Example 1: Imagery from FSMI projected onto a map, displayed as a movie

```python
from datetime import datetime

import asilib.asi
import asilib.map

time_range = (datetime(2023, 3, 24, 6, 0), datetime(2023, 3, 24, 6, 30))
location_code = 'FSMI'

asi = asilib.asi.themis(location_code, time_range=time_range, alt=110)

lat_bounds = (asi.meta['lat'] - 7, asi.meta['lat'] + 7)
lon_bounds = (asi.meta['lon'] - 20, asi.meta['lon'] + 20)
ax = asilib.map.create_map(lon_bounds=lon_bounds, lat_bounds=lat_bounds)

asi.animate_map(ax=ax)

print(f'Animation saved in {asilib.config["ASI_DATA_DIR"] / "animations" / asi.animation_name}')
```

Example 2: Fisheye animation of FSMI ASI data

```python
from datetime import datetime

import asilib.asi

location_code = 'FSMI'
time_range = (datetime(2023, 3, 24, 6, 0), datetime(2023, 3, 24, 6, 30))
asi = asilib.asi.themis(location_code, time_range=time_range)
asi.animate_fisheye()

print(f'Animation saved in {asilib.config["ASI_DATA_DIR"] / "animations" / asi.animation_name}')
```
